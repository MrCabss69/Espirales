Coordenadas polares: 


> $ x = r * cos(\theta)  $ 

> $ y = r * sen(\theta)) $

Siendo r el módulo o la distancia del punto al origen, y $\theta$ el ángulo del vector

```python
def to_cart(r,t):
  return r*math.cos(t) , r*math.sin(t)
```

```python
def en_polares(x,y):
  r = math.sqrt(x**2 + y ** 2)
  t = np.arctan(y/x)
  return r,t
```

La espiral áurea es una espiral logarítmica o espiral de crecimineto. Éstas espirales de forma general se definen como:

$ r = ae^{bθ} → \theta = log_{b}{(r/a)} $  

con   $ b = \phi $, siendo $a$ una constante y $e$ la base del logaritmo natural (2.71..). Además de la ecuación general, a espiral áurea cumple la particularidad de tener un factor de crecimiento $\phi$.

Esto significa que cuando el ángulo es $ \frac{π}{2} → e^{b\theta} = \phi $

```python
g = 1.618 # phi
b = math.log(g)/(math.pi/2) # valor de exp. para espiral áurea
limite = 1000
```

```python
def get_golden(a):
  global limite,b
  xx  = np.linspace(0.001,limite,limite*10)
  x,y,z = [],[], []
  for r in xx:
    # calculamos theta
    t = math.log(r/a)/b
    # convertimos los puntos a cartesianos
    carts = to_cart(r,t)
    x += [carts[0]]
    y += [carts[1]]
    z += [r]
  return tuple(map(np.array,[x,y,z]))
```

```python
import matplotlib.pyplot as plt
x,y,_ = get_golden(10000)
plt.plot(x,y,'red') 
plt.plot(-np.array(x),-np.array(y))
plt.xlim(-limite,limite)
plt.ylim(-limite,limite)
plt.show()
```

```python
import matplotlib.pyplot as plt
# set up a figure twice as wide as it is tall
fig = plt.figure(figsize=plt.figaspect(0.25))

# =============
# First subplot
# =============
# set up the axes for the first plot
ax = fig.add_subplot(1, 2, 1, projection='3d')

# Second subplot
# =============
# set up the axes for the 2nd plot
ax2 = fig.add_subplot(1,  2, 2)
for a in [ 10**i for i in range(90)]:
  x,y,z = get_golden(a)
  ax.plot3D(x,y,z)
  ax2.plot(-x,y)
plt.show()
```
